SEL 1: Memuat Dataset Titanic & Set Up Library

Instruksi: Mengimpor library Matplotlib dan Seaborn, memuat dataset external Titanic, serta mengatur tema visualisasi agar seragam.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Mengatur tema dasar visualisasi dengan Seaborn
sns.set_theme(style="darkgrid")

# 2. Memuat dataset external Titanic bawaan Seaborn sesuai request kustom
df = sns.load_dataset('titanic')

# Menampilkan informasi awal data untuk memastikan sukses ter-load
print(f"Dataset Titanic Berhasil Dimuat! Ukuran: {df.shape} (Baris, Kolom)\n")
print(df.head())
```

SEL 2: Matplotlib Dasar – Bar Chart Kelas Penumpang

Instruksi: Menggunakan fungsi dasar Matplotlib (plt.subplots) untuk menampilkan jumlah total penumpang di setiap kelas tiket (pclass).

```python
# 1. Menghitung ringkasan data jumlah penumpang per kelas
pclass_counts = df['pclass'].value_counts().sort_index()

# 2. Membuat komponen Figure & Axes Matplotlib
fig, ax = plt.subplots(figsize=(7, 4.5))

# 3. Menggambar Bar Chart
ax.bar(pclass_counts.index.astype(str), pclass_counts.values, color=['#4C72B0', '#DD8452', '#55A868'])

# 4. Mengatur Label dan Judul
ax.set_title('Jumlah Penumpang Berdasarkan Kelas Tiket (Pclass)', fontsize=13, pad=12)
ax.set_xlabel('Kelas Tiket (Pclass)')
ax.set_ylabel('Jumlah Penumpang')

plt.tight_layout()
plt.show()
```

SEL 3: Seaborn Statistik – Scatter Plot Hubungan Umur vs Tarif

Instruksi: Menggunakan Seaborn untuk menggambar grafik bivariat guna melihat hubungan korelasi sebaran titik antara variabel age (umur) dengan fare (tarif tiket) dipisah berdasarkan variabel survived.

```python
plt.figure(figsize=(8, 5))

# Membuat scatter plot hubungan umur vs tarif tiket
sns.scatterplot(data=df, x='age', y='fare', hue='survived', alpha=0.7, palette='Set1')

plt.title('Hubungan Antara Umur dan Tarif Tiket Penumpang', fontsize=13, pad=12)
plt.xlabel('Umur Penumpang')
plt.ylabel('Tarif Tiket (Fare)')
plt.legend(title='Selamat? (1=Ya, 0=Tidak)')
plt.show()
```

SEL 4: Seaborn Univariat – Boxplot Distribusi Kelompok Tarif

Instruksi: Membuat diagram kotak (Boxplot) untuk mendeteksi sebaran nilai kuartil, median, serta sebaran titik data pencilan (outliers) dari variabel tarif tiket (fare) berdasarkan pelabuhan keberangkatan (embarked).

```python
plt.figure(figsize=(8, 5))

# Membuat diagram Boxplot distribusi tarif per pelabuhan
sns.boxplot(data=df, x='embarked', y='fare', hue='embarked', palette='pastel', legend=False)

plt.title('Distribusi Tarif Tiket Berdasarkan Pelabuhan Keberangkatan', fontsize=13, pad=12)
plt.xlabel('Pelabuhan Keberangkatan (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Tarif Tiket (Fare)')
# Membatasi sumbu y agar boxplot utama terlihat jelas dari outlier ekstrem
plt.ylim(0, 150)
plt.show()
```
